```python
def view_books(books):
    print("\nAll Books:")
    for i, book in enumerate(books, start=1):
        status = "Available" if book["available"] else "Checked Out"
        print(f'{i}. "{book["title"]}" by {book["author"]} ({status})')

def search_book(books):
    search_term = input("\nEnter the title or author to search for: ").lower()
    found_books = [book for book in books if search_term in book["title"].lower() or search_term in book["author"].lower()]
    
    if found_books:
        print("\nSearch Results:")
        for i, book in enumerate(found_books, start=1):
            status = "Available" if book["available"] else "Checked Out"
            print(f'{i}. "{book["title"]}" by {book["author"]} ({status})')
    else:
        print("\nNo books found.")

def borrow_book(books):
    view_books(books)
    choice = int(input("\nEnter the number of the book you want to borrow: ")) - 1
    
    if 0 <= choice < len(books) and books[choice]["available"]:
        books[choice]["available"] = False
        print(f'\nYou have successfully borrowed "{books[choice]["title"]}".')
    else:
        print("\nInvalid choice or the book is already checked out.")

def return_book(books):
    view_books(books)
    choice = int(input("\nEnter the number of the book you want to return: ")) - 1
    
    if 0 <= choice < len(books) and not books[choice]["available"]:
        books[choice]["available"] = True
        print(f'\nYou have successfully returned "{books[choice]["title"]}".')
    else:
        print("\nInvalid choice or the book is already available.")

def view_users(users):
    print("\nAll Users:")
    for i, user in enumerate(users, start=1):
        print(f'{i}. {user["name"]}')

def menu():
    books = [
        {"title": "To Kill a Mockingbird", "author": "Harper Lee", "available": True},
        {"title": "1984", "author": "George Orwell", "available": False},
        {"title": "The Great Gatsby", "author": "F. Scott Fitzgerald", "available": True}
    ]

    users = [
        {"name": "John Doe"},
        {"name": "Jane Smith"},
        {"name": "Alice Johnson"}
    ]

    while True:
        print("\nPlease choose an option:")
        print("1. View all books")
        print("2. Search for a book")
        print("3. Borrow a book")
        print("4. Return a book")
        print("5. View all users")
        print("6. Exit")

        choice = input("\nEnter your choice (1-6): ")

        if choice == '1':
            view_books(books)
        elif choice == '2':
            search_book(books)
        elif choice == '3':
            borrow_book(books)
        elif choice == '4':
            return_book(books)
        elif choice == '5':
            view_users(users)
        elif choice == '6':
            print("\nExiting the system.")
            break
        else:
            print("\nInvalid choice, please try again.")

# Instead of __main__, simply call the menu function directly
menu()
```
